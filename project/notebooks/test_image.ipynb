```python
import os
import torchaudio
torchaudio.set_audio_backend("sox_io")
from torchaudio.transforms import MelSpectrogram
from torchvision import transforms
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path


sound_path = Path("..","data", "raw_data","french","anger","anger_1_7d31c24a-8cb1-11eb-8203-b216f6aef6a6.wav")
waveform, sample_rate = torchaudio.load(str(sound_path))
image_tensor = MelSpectrogram(sample_rate=sample_rate, n_mels=52)(waveform)
im = transforms.ToPILImage("F")(image_tensor)

plt.figure()
plt.subplot(3,1,1)
plt.plot(np.squeeze(waveform))
plt.subplot(3,1,2)
plt.imshow(np.squeeze(image_tensor))
plt.subplot(3,1,3)
plt.imshow(im)
```
