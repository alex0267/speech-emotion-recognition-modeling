```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
from IPython.display import Audio, IFrame, display
from pathlib import Path
from functools import partial
import matplotlib.pyplot as plt
import librosa
import librosa.display as lds
import numpy as np

import warnings
warnings.filterwarnings("ignore")

SOUNDS_ROOT = Path("../data/raw_data/french")
PICS_ROOT   = Path("../data/test")

def _filter(suffix,item):
    return str(item).endswith(suffix)
    
def _transform(item):
    return str(item)
    
sounds_paths = list(map(_transform,filter(partial(_filter,".wav"), SOUNDS_ROOT.glob("**/*.wav"))))
pics_paths = list(map(_transform,filter(partial(_filter,".jpg"),PICS_ROOT.glob("**/*.jpg"))))
test_sounds_paths = list(map(_transform,filter(partial(_filter,".wav"),PICS_ROOT.glob("**/*.wav"))))


                  

```

```python
def display_file(file,prefix=""):
    y,sr = librosa.load(file, mono=True, sr=48000, offset=0) # load the audio file
    display(Audio(y,rate=sr)) #display
    fig, ax = plt.subplots()
    img=lds.waveplot(y,sr=sr, x_axis=f"time", ax=ax)
    ax.set_title(f"{prefix} data")
    
    fig, ax = plt.subplots()
    S1 = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=52)
    D1 = librosa.power_to_db(S1, ref=np.max)
    lds.specshow(D1, x_axis=f"time", y_axis='mel',ax=ax);
    fig.colorbar(img, ax=ax)
    ax.set_title(f"{prefix} mel spectrogram")
    
    fig, ax = plt.subplots()
    rms = librosa.feature.rms(y=y,frame_length=1024, hop_length=512)
    plt.semilogy(rms.T)
    ax.set_title(f"{prefix} Energy")
    
@interact
def show_sound(file=sounds_paths):
    from pathlib import Path
    display_file(file)
    file_name = Path(file).stem
    test_sounds = list(PICS_ROOT.glob(f"**/*{file_name}.wav"))
    if len(test_sounds):
        display_file(str(test_sounds[0]),"transformed ")
    
```
